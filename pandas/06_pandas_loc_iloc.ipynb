```python
import pandas as pd
```

```python
df = pd.read_csv("../data/superstore_data.csv",parse_dates=['Dt_Customer'])
df.head()
```

```python
df.set_index('Dt_Customer',inplace=True)
df.head()
```

## loc method : loc method used to access group by rows/column by index labels

- df.loc[label] : retrieve row(s) for particular index label
- df.loc[[label1,label2,..]] : retrieve group of rows corresponding to passed list of labels
- df.loc[start : end] : loc parameters can also be passed in form of list slice to retrieve data(range is inclusive)

### Using loc to extract records for single value of index

```python
df.loc['2014-08-04']
```

### passing multiple index values to loc method

```python
df.loc[['2014-06-15','2014-08-04']]
```

### using slice operator with loc method

```python
df.reset_index(inplace=True,drop=False)
df.set_index('Id',inplace=True)
df.sort_index(inplace=True)
df.loc[1:20]
```

```python
df.loc[20]
```

## iloc method : used to access group of rows by numerical index value

```python
df.reset_index(inplace=True,drop=False)
df.iloc[0]
```

```python
df.iloc[[3,6,9]]
```

```python
df.iloc[10:14]
```

### passing second arguement(columns) to loc/iloc method

```python
columns_subset = ['Id','Year_Birth','Education','Income']
df.set_index('Dt_Customer',inplace=True)
df.loc['2013-09-18',columns_subset]
```

```python
df.iloc[100:106,2:7]
```

```python
df.iloc[[6,8,45,88],[1,6,8,9]]
```
