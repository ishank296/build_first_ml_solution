### The pivot() method

```python
salesmen = pd.read_csv("../data/salesmen.csv",parse_dates=['Date'])
```

```python
salesmen.head()
```

```python
salesmen.pivot(index='Date',columns='Salesman',values='Revenue')
```

### The .pivot_table() method

```python
salesmen.pivot_table(values='Revenue',aggfunc='sum',index='Salesman')
```

```python
world_stats.pivot_table(values='GDP',aggfunc='mean',columns='country',index='year').head()
```

```python
world_stats.pivot_table(values='GDP',aggfunc='sum',index=['country','year'])
```
