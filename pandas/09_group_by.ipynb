```python
import pandas as pd
```

```python
fortune_df = pd.read_csv("../data/fortune1000.csv",index_col='Rank')
fortune_df.head(3)
```

```python
group_by_sector=fortune_df.groupby('Sector')
```

```python
len(group_by_sector)
```

### .size() method returns output same as .value_count() called on actual dataframe

```python
group_by_sector.size()
```

```python
fortune_df['Sector'].value_counts()
```

```python
group_by_sector.first().head(5)
```

### .get_group() method

```python
group_by_sector.get_group('Energy')
```

### Calling aggregate methods e.g. max(),sum(),mean() on grouped dataframe


```python
group_by_sector.sum(numeric_only=True).head()
```

```python
group_by_sector.max(['Revenue','Profits']).head()
```

```python
group_by_sector.max().tail(5)
```

```python
group_by_sector['Profits'].sum()
```

```python
group_by_sector['Employees'].mean().head()
```

```python
group_by_sector[['Employees','Revenue']].mean().head()
```

### Grouping by multiple columns

```python
grp_by_sector_industries = fortune_df.groupby(['Sector','Industry'])
```

```python
grp_by_sector_industries.sum(numeric_only=True)
```

```python
grp_by_sector_industries.get_group(('Apparel','Apparel'))
```

```python
grp_by_sector_industries['Employees'].mean()
```

### The .agg() method

```python
group_by_sector.agg({'Revenue':'sum',
                     'Profits':'sum',
                     'Employees':'mean'}).head(10)
```
